```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
path1 = r'E:\netData\MAGzip\MAG Nobel\2 Nobelpapers\1M NobelCitingpapers.zip'
path2 = r'E:\netData\MAGzip\MAG Nobel\2 Nobelpapers\712 Nobelpapers.zip'
```

```python
f1 = pd.read_csv(path1)
f2 = pd.read_csv(path2)
```

# fig 1

```python
id = f2['paperid'].values[0]
x1 = f1.loc[(f1['Nobel'] == id) & (f1['type'] == 'DC'), 'paperyear'].values
x2 = f1.loc[(f1['Nobel'] == id) & (f1['type'] == 'CC'), 'paperyear'].values
y1 = f1.loc[(f1['Nobel'] == id) & (f1['type'] == 'DC'), 'cit'].values
y2 = f1.loc[(f1['Nobel'] == id) & (f1['type'] == 'CC'), 'cit'].values
y2 = [-i for i in y2]
```

```python
f, ax = plt.subplots(figsize=(10, 8))

ax.stem(x1, y1, linefmt='C0-.', markerfmt='C0o', basefmt=' ' , label = 'Disruptive Citing Papers (DCP)')
ax.stem(x2, y2, linefmt='C1-.', markerfmt='C1D', basefmt=' ' , label = 'Consolidating Citing Papers (CCP)')

ax.axhline(0,ls = "-",color = '#283747')
# ax.yaxis.grid(True)
plt.ylim(-620,620)
plt.xticks(size = 28)
plt.yticks([-600,-300,0,300,600],[600,300,0,300,600],size = 28)
plt.legend(fontsize = 25)
plt.xlabel(r'year',fontsize = 32)
plt.ylabel(r'Impact (Citation)',fontsize = 32)
plt.title(r'Citing Pubs',fontsize = 32)
plt.savefig(r"G:\A J YANG PAPER\MAG analysis\papers Nobel\JASIST Disruptive citation of Nobel papers\fig\fig 1\a1.pdf",bbox_inches = 'tight')
```
